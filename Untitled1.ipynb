```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax=plt.subplots(1,1)
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27,63,71,90,92,96,32,37,40])
plt.hist(a)
ax.set_xlabel('marks')
ax.set_ylabel('no of students')
plt.show()
```
